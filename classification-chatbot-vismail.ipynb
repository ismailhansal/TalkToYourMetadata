```python
%%writefile app.py
import streamlit as st
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
import pandas as pd
import json
from IPython.display import Markdown
import bitsandbytes
import accelerate
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, AutoConfig, pipeline
from langchain.document_loaders import TextLoader
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain_huggingface import HuggingFacePipeline
from langchain.chains import LLMChain
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.pydantic_v1 import BaseModel, Field
from langchain.schema.output_parser import StrOutputParser
from langchain.text_splitter import CharacterTextSplitter
from glob import glob
import requests
import matplotlib.pyplot as plt
import re
import sqlite3
import hashlib



def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

# Créer une base de données SQLite et une table pour les utilisateurs
conn = sqlite3.connect('users.db')
c = conn.cursor()
c.execute('''
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        username TEXT UNIQUE,
        password TEXT
    )
''')
conn.commit()

# Fonction pour ajouter un utilisateur
def add_user(username, password):
    try:
        c.execute('INSERT INTO users (username, password) VALUES (?, ?)', (username, hash_password(password)))
        conn.commit()
        st.success("Utilisateur ajouté avec succès!")
    except sqlite3.IntegrityError:
        st.error("Le nom d'utilisateur existe déjà. Veuillez en choisir un autre.")

# Fonction pour vérifier les informations de connexion
def authenticate(username, password):
    c.execute('SELECT * FROM users WHERE username = ? AND password = ?', (username, hash_password(password)))
    return c.fetchone()

# Fonction pour supprimer un utilisateur
def delete_user(username):
    c.execute('DELETE FROM users WHERE username = ?', (username,))
    conn.commit()
    st.success(f"Utilisateur {username} supprimé avec succès!")

# Fonction pour obtenir tous les utilisateurs
def get_all_users():
    c.execute('SELECT * FROM users')
    return c.fetchall()

# Page d'inscription
def register():
    st.title("Inscription")

    with st.form(key='register_form'):
        username = st.text_input("Nom d'utilisateur")
        password = st.text_input("Mot de passe", type="password")
        submit_button = st.form_submit_button(label="S'inscrire")

        if submit_button:
            if username and password:
                try:
                    add_user(username, password)
                    st.success("Inscription réussie! Veuillez vous connecter.")
                    st.session_state['show_login'] = True  # Rediriger vers la page de connexion après l'inscription
                    st.session_state['show_register'] = False
                except sqlite3.IntegrityError:
                    st.error("Le nom d'utilisateur existe déjà. Veuillez en choisir un autre.")
            else:
                st.error("Veuillez entrer un nom d'utilisateur et un mot de passe.")
    if st.button("Se connecter"):
        st.session_state["show_register"] = False
        st.session_state["show_login"] = True

# Page de connexion
def login():
    st.title("Connexion")

    with st.form(key='login_form'):
        username = st.text_input("Nom d'utilisateur")
        password = st.text_input("Mot de passe", type="password")
        submit_button = st.form_submit_button(label="Se connecter")

        if submit_button:
            user = authenticate(username, password)
            if user:
                st.session_state["authenticated"] = True
                st.success("Connexion réussie!")
            else:
                st.error("Nom d'utilisateur ou mot de passe incorrect")

    if st.button("Créer un compte"):
        st.session_state["show_register"] = True
        st.session_state["show_login"] = False

# Page d'administration
def admin_page():
    st.title("Gestion des Utilisateurs")

    tab1, tab2 = st.tabs(["Ajouter un Utilisateur", "Supprimer un Utilisateur"])

    with tab1:
        st.subheader("Ajouter un Utilisateur")
        with st.form(key='add_user_form'):
            username = st.text_input("Nom d'utilisateur")
            password = st.text_input("Mot de passe", type="password")
            submit_button = st.form_submit_button(label="Ajouter")

            if submit_button:
                if username and password:
                    add_user(username, password)
                else:
                    st.error("Veuillez entrer un nom d'utilisateur et un mot de passe.")

    with tab2:
        st.subheader("Supprimer un Utilisateur")
        users = get_all_users()
        user_list = [user[1] for user in users]  # Récupère les noms d'utilisateur
        selected_user = st.selectbox("Choisissez un utilisateur à supprimer", user_list)
        delete_button = st.button("Supprimer l'utilisateur")

        if delete_button:
            if selected_user:
                delete_user(selected_user)
            else:
                st.error("Veuillez sélectionner un utilisateur.")

    # Afficher tous les utilisateurs
    st.subheader("Liste des Utilisateurs")
    users = get_all_users()
    for user in users:
        st.write(f"ID: {user[0]}, Nom d'utilisateur: {user[1]}")

# Initialisation des états de session
if "authenticated" not in st.session_state:
    st.session_state["authenticated"] = False

if "show_register" not in st.session_state:
    st.session_state["show_register"] = False

if "show_login" not in st.session_state:
    st.session_state["show_login"] = True

if "show_admin" not in st.session_state:
    st.session_state["show_admin"] = False

# Afficher la page appropriée en fonction de l'état de session
if not st.session_state["authenticated"]:
    if st.session_state["show_register"]:
        register()
    else:
        login()
else:
    if st.session_state["show_admin"]:
        admin_page()
    else:
        st.title("Talk to your Metadata")
        

    # Uploader les fichiers
    csv_file = st.file_uploader("Choose csv file to classify")
    txt_file = st.file_uploader("Choose text file to classify", type="txt")

    run = st.button("Classify tables in csv file")
    run_txt = st.button("Classify Document")

    r1 = st.container()
    r2 = st.container()

    # Fonction de requête au modèle
    @st.cache_data
    def query_model(
            system_message,
            user_message,
            temperature=0.7,
            max_length=1024
            ):
        prompt_template = PromptTemplate(template=system_message, input_variables=["context"])
        
        def format_docs(docs):
            return "\n\n".join([d.page_content for d in docs])
        
        chain = {"context": retriever | format_docs} | prompt_template
        system_message = chain.invoke(input="I'm a context").text
        
        messages = [
            {"role": "system", "content": system_message},
            {"role": "user", "content": user_message},
        ]
        
        
        
        prompt = pipeline.tokenizer.apply_chat_template(
            messages, 
            tokenize=False, 
            add_generation_prompt=True
        )
        
        
        
        print("Final Prompt:")
        print(prompt)
        
        terminators = [
            pipeline.tokenizer.eos_token_id,
            pipeline.tokenizer.convert_tokens_to_ids("")
        ]
        
        sequences = pipeline(
            prompt,
            do_sample=True,
            top_p=0.9,
            temperature=temperature,
            eos_token_id=terminators[0] if isinstance(terminators, list) else terminators,
            max_new_tokens=max_length,
            return_full_text=False,
            pad_token_id=pipeline.model.config.pad_token_id if pipeline.model.config.pad_token_id is not None else -1
        )

        answer = sequences[0]['generated_text']
        return answer

    @st.cache_resource
    def init_model(model):
        pipeline = transformers.pipeline(
            "text-generation",
            model=model,
            torch_dtype=torch.float16,
            device_map="auto",
        )
        return pipeline

    @st.cache_data
    def generate_chunks(paper_paths):
        paper_paths = glob(paper_paths)
        pages = []

        for path in paper_paths:
            try:
                loader = TextLoader(path)
                doc = loader.load()

                # Chunk text
                text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=0)
                chunked_documents = text_splitter.split_documents(doc)

                pages.extend(chunked_documents)
            except Exception as e:
                print(f'Skipping {path} due to error: {e}')
        return pages

    
    @st.cache_resource
    def retriever_gen(_pages):
        db = FAISS.from_documents(
            _pages,
            HuggingFaceEmbeddings(model_name='sentence-transformers/all-mpnet-base-v2')
        )

        retriever = db.as_retriever(
            search_type="similarity",
            search_kwargs={'k': 4}
        )
        return retriever

    pipeline = init_model("/kaggle/input/llama-3.1/transformers/8b-instruct/1")
    pages = generate_chunks("/kaggle/working/docs/*.txt")
    retriever = retriever_gen(pages)

    system_message_tables = """
    **Classification des Données Sensibles**

    **Objectif** : Classer les colonnes de données en fonction de leur sensibilité pour garantir la sécurité et la confidentialité des informations.

    **Critères de Classification** :

    * Impact Financier
    * Impact Conformité
    * Impact Réputation
    * Impact Opérationnel

    **Niveaux de Sensibilité** :

    1. Très Secret
    2. Secret
    3. Confidentiel
    4. Public
    5. Restreint

    **Règles de Classification** :

    * Choisir le niveau de classification selon la valeur maximale des critères.
    * Si plusieurs critères ont la même valeur maximale, choisir le niveau de classification le plus élevé.
    * Fournir des justifications pour chaque niveau de sensibilité attribué, en se basant sur le contenu du document.

    **Réponse Attendue** :

    **Classification des Colonnes**

    * Colonne1 : [Niveau de Sensibilité] ([Justification])
    * Colonne2 : [Niveau de Sensibilité] ([Justification])
    * ...

    **Occurrences de chaque Niveau de Sensibilité**

    * Occurence Très Secret : [Nombre]
    * Occurence Secret : [Nombre]
    * Occurence Confidentiel : [Nombre]
    * Occurence Public : [Nombre]
    * Occurence Restreint : [Nombre]
    """

    system_message_docs = """
    **Classification des Documents Sensibles**

    **Objectif :** Classer les documents en fonction de leur sensibilité pour garantir la sécurité et la confidentialité des informations.

    **Critères de Classification :**
    - Impact Financier
    - Impact Conformité
    - Impact Réputation
    - Impact Opérationnel

    **Niveaux de Sensibilité :**
    1. Très Secret
    2. Secret
    3. Confidentiel
    4. Restreint
    5. Public

    **Règles de Classification :**
    - Choisir le niveau de classification en fonction des critères de sensibilité.
    - Fournir des justifications pour chaque niveau de sensibilité attribué, en se basant sur le contenu du document.

    **Réponse Attendue :**
    - **Classification :** Niveau de sensibilité attribué au document.
    - **Justifications :** Liste des éléments ou phrases du document qui ont conduit à cette classification. Mentionner les critères de sensibilité observés.

    Note : veuillez respecter la syntaxe concernant les occurences

    **Occurrences de chaque Niveau de Sensibilité :**
    - **Occurence Très Secret :** Nombre de mentions d'informations très sensibles. (justification)
    - **Occurence Secret :** Nombre de mentions d'informations secrètes. (justification)
    - **Occurence Confidentiel :** Nombre de mentions d'informations confidentielles. (justification)
    - **Occurence Restreint :** Nombre de mentions d'informations restreintes. (justification)
    - **Occurence Public :** Nombre de mentions d'informations publiques. (justification)

    **Note :** Les justifications doivent se baser sur des éléments concrets du document. Les occurrences de chaque niveau de sensibilité seront utilisées pour créer une visualisation de la répartition des données sensibles.
    """

    
    
    if run and csv_file:
        dataframe = pd.read_csv(csv_file)
        table = csv_file.name
        user_message = f"Classify the columns of the following table named {table} based on their sensitivity level. The table has the following columns: {', '.join(dataframe.columns)}."

        response = query_model(
            system_message_tables,
            user_message=user_message,
            temperature=0.7,
            max_length=500
        )

        # Initialiser un dictionnaire pour les occurrences
        occurrences = {
            "Très Secret": 0,
            "Secret": 0,
            "Confidentiel": 0,
            "Public": 0,
            "Restreint": 0
        }

        # Extraire les informations de classification
        for line in response.split("\n"):
            if "Occurence Très Secret" in line:
                occurrences["Très Secret"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Secret" in line:
                occurrences["Secret"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Confidentiel" in line:
                occurrences["Confidentiel"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Restreint" in line:
                occurrences["Restreint"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Public" in line:
                occurrences["Public"] = int(re.search(r"(\d+)", line).group(1))

        # Afficher le résultat
        with r1:
            st.subheader("Résultat de la Classification des Colonnes")
            st.write(response)

            # Créer un graphique pour visualiser les occurrences
            fig, ax = plt.subplots()
            ax.bar(occurrences.keys(), occurrences.values(), color='skyblue')
            ax.set_xlabel("Niveau de Sensibilité")
            ax.set_ylabel("Nombre d'Occurrences")
            ax.set_title("Répartition des Occurrences par Niveau de Sensibilité")
            plt.xticks(rotation=45)
            st.pyplot(fig)

    if run_txt and txt_file:
        document = txt_file.read().decode("utf-8")
        user_message_docs = f"Classify the following document based on its sensitivity level. Provide justification for each classification level: {document}"

        response_docs = query_model(
            system_message_docs,
            user_message=user_message_docs,
            temperature=0.7,
            max_length=500
        )

        # Initialiser un dictionnaire pour les occurrences
        occurrences_docs = {
            "Très Secret": 0,
            "Secret": 0,
            "Confidentiel": 0,
            "Public": 0,
            "Restreint": 0
        }

        # Extraire les informations de classification
        for line in response_docs.split("\n"):
            if "Occurence Très Secret" in line:
                occurrences_docs["Très Secret"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Secret" in line:
                occurrences_docs["Secret"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Confidentiel" in line:
                occurrences_docs["Confidentiel"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Restreint" in line:
                occurrences_docs["Restreint"] = int(re.search(r"(\d+)", line).group(1))
            elif "Occurence Public" in line:
                occurrences_docs["Public"] = int(re.search(r"(\d+)", line).group(1))

        # Afficher le résultat
        with r2:
            st.subheader("Résultat de la Classification des Documents")
            st.write(response_docs)

            # Créer un graphique pour visualiser les occurrences
            fig, ax = plt.subplots()
            ax.bar(occurrences_docs.keys(), occurrences_docs.values(), color='lightcoral')
            ax.set_xlabel("Niveau de Sensibilité")
            ax.set_ylabel("Nombre d'Occurrences")
            ax.set_title("Répartition des Occurrences par Niveau de Sensibilité")
            plt.xticks(rotation=45)
            st.pyplot(fig)
```
